```python
#=========================Task 1: Create the Book Class========================


#====Task 1: Create the Book Class=====

class Book():

 def __init__ (self, title, author, isbn, is_checked_out=False):

  self.title = title
  self.author = author
  self.isbn = isbn
  self.is_checked_out = is_checked_out


#============================Task 2: Implement Instance Method  ===============

# 1: ===display_info====

 def display_info (self):
  if self.is_checked_out:
    status = "Checked Out"
  else:
    status = "Available"
  return(f"Book= {self.title} \nAuthor = {self.author} \nISBN # = {self.isbn} \nReturn= {status}")

# 2: ===checked_out====
 def check_out (self):
    if self.is_checked_out:
      print(f"{self.title} is already checked out")
    else:
      self.is_checked_out=True
      print(f"{self.title} has been checked out")

# 3: ===CHECK_IN====

 def check_in (self):
   if self.is_checked_out:
       self.is_checked_out=False
       print(f"{self.title} has been checked in ")
   else:
    print(f"{self.title} is already available ")


#=========================Task 3: Library class  ============================

class Library():

 def __init__ (self,library_name):

    self.library_name=library_name
    self.books=[]

#====  Implement Instance Method  =====

# 1: ===add_book====

 def add_book(self, book_object):

   self.books.append(book_object)
   print(f"{book_object.title} added to {self.library_name}")

# 2: ===find_book_by_title====

 def find_book_by_title(self, title):

   for book in self.books:
      if book.title == title:
       return book
   print(f"Book with title {title} not found ")
   return None

# 3: ===check_out_book====

 def check_out_book(self, title):
   book = self.find_book_by_title(title)
   if book:
    book.check_out()

# 4: ===check_in_book====

 def check_in_book(self, title):
   book = self.find_book_by_title(title)
   if book:
    book.check_in()

# 5: =====list_available_books====

 def list_available_books(self):
   print(f"Available books in {self.library_name}:")

   available_books = []
   for book in self.books:
        if not book.is_checked_out:
            available_books.append(book)

   if available_books:
        for book in available_books:
            print(book.display_info())
   else:
        print("No books are currently available")

# # 6: =====list_all_books====

 def list_all_books(self):
   print(f"All books in {self.library_name}. ")

   if self.books:
    for book in self.books:
        print(book.display_info())
   else:
        print("No books are currently available")





#========================Task 4: Test Your System=========================

#======Create a Library
my_library = Library("Downtown Coding Library")


#=============Create Books

book1 = Book("Python for Beginners", "John Coder", "1234567890")
book2 = Book("Advanced OOP Concepts", "Jane Hacker", "0987654321")
book3 = Book("Data Structures in Python", "Alan Turing Jr.", "1122334455")
print("-" * 40 )

#=========== Add Books to Library

my_library.add_book(book1)
my_library.add_book(book2)
my_library.add_book(book3)
print("-" * 20)



#============ List all books
my_library.list_all_books()
print("-" * 20)


#================ Check out a book
my_library.check_out_book("Python for Beginners")
print("-" * 20)

#============== Try checking out the same book again
my_library.check_out_book("Python for Beginners")
print("-" * 20)

# 8. List available books again
my_library.list_available_books()
print("-" * 20)


# 9. Check in a book
my_library.check_in_book("Python for Beginners")
print("-" * 20)

# 10. Try checking in an already available book


my_library.check_in_book("Advanced OOP Concepts") # This book was never checked out
print("-" * 20)



# 11. List all books to see final status
my_library.list_all_books()
print("-" * 20)


# 12. Find a book by title


found_book = my_library.find_book_by_title("Advanced OOP Concepts")
if found_book:
 print(f"Found book: {found_book.title} by {found_book.author}")
print("-" * 20)


# 13. Try finding a non-existent book
my_library.find_book_by_title("The Art of Assembly")
print("-" * 20)

```
